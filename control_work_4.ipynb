Реализуйте класс очередь на основе списка и на основе deque. Сравните время и затраты памяти на каждую структуру при последовательном добавлении и извлечении элементов.

```python
from collections import deque
import matplotlib.pyplot as plt
import time
import sys

class ListQueue:
    def __init__(self):
        self.queue = []

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if len(self.queue) < 1:
            return None
        return self.queue.pop(0)
    
    def size(self):
        return len(self.queue)


class DequeQueue:
    def __init__(self):
        self.queue = deque()

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if len(self.queue) < 1:
            return None
        return self.queue.popleft()

    def size(self):
        return len(self.queue)


def measure_performance(queue_class, operations):
    start_time = time.time()
    memory_before = sys.getsizeof(queue_class.queue)

    for i in range(operations):
        queue_class.enqueue(i)

    for i in range(operations):
        queue_class.dequeue()

    end_time = time.time()
    memory_after = sys.getsizeof(queue_class.queue)

    time_taken = end_time - start_time
    memory_used = memory_after - memory_before

    return time_taken, memory_used


operations = 100000
list_queue = ListQueue()
deque_queue = DequeQueue()

list_performance = measure_performance(list_queue, operations)
deque_performance = measure_performance(deque_queue, operations)

list_performance, deque_performance
```

```python
def measure_individual_operations_and_memory_queue(queue_class, operations):
    enqueue_times = []
    dequeue_times = []
    enqueue_memory = []
    dequeue_memory = []

    for i in range(operations):
        start_time = time.time()
        queue_class.enqueue(i)
        end_time = time.time()
        enqueue_times.append(end_time - start_time)
        enqueue_memory.append(sys.getsizeof(queue_class.queue))

    for i in range(operations):
        start_time = time.time()
        queue_class.dequeue()
        end_time = time.time()
        dequeue_times.append(end_time - start_time)
        dequeue_memory.append(sys.getsizeof(queue_class.queue))

    return enqueue_times, dequeue_times, enqueue_memory, dequeue_memory

operations = 100000


list_enqueue_times, list_dequeue_times, list_enqueue_memory, list_dequeue_memory = measure_individual_operations_and_memory_queue(list_queue, operations)
deque_enqueue_times, deque_dequeue_times, deque_enqueue_memory, deque_dequeue_memory = measure_individual_operations_and_memory_queue(deque_queue, operations)


plt.figure(figsize=(12, 12))

# Время операций добавления в очередь
plt.subplot(2, 2, 1)
plt.plot(range(operations), list_enqueue_times, color='blue', label='Время добавления в ListQueue')
plt.plot(range(operations), deque_enqueue_times, color='red', label='Время добавления в DequeQueue')
plt.title('Время операций добавления')
plt.xlabel('Номер операции')
plt.ylabel('Время (секунды)')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(range(operations), list_dequeue_times, color='blue', label='Время извлечения из ListQueue')
plt.plot(range(operations), deque_dequeue_times, color='red', label='Время извлечения из DequeQueue')
plt.title('Время операций извлечения')
plt.xlabel('Номер операции')
plt.ylabel('Время (секунды)')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(range(operations), list_enqueue_memory, color='blue', label='Использование памяти при добавлении в ListQueue')
plt.plot(range(operations), deque_enqueue_memory, color='red', label='Использование памяти при добавлении в DequeQueue')
plt.title('Использование памяти во время добавления')
plt.xlabel('Номер операции')
plt.ylabel('Использование памяти (байты)')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(range(operations), list_dequeue_memory, color='blue', label='Использование памяти при извлечении из ListQueue')
plt.plot(range(operations), deque_dequeue_memory, color='red', label='Использование памяти при извлечении из DequeQueue')
plt.title('Использование памяти во время извлечения')
plt.xlabel('Номер операции')
plt.ylabel('Использование памяти (байты)')
plt.legend()

plt.tight_layout()
plt.show()


```

Реализуйте класс стек на основе списка и на основе deque. Сравните время и затраты памяти на каждую структуру при последовательном добавлении и извлечении элементов.

```python
class ListStack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if len(self.stack) < 1:
            return None
        return self.stack.pop()

    def size(self):
        return len(self.stack)


class DequeStack:
    def __init__(self):
        self.stack = deque()

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if len(self.stack) < 1:
            return None
        return self.stack.pop()

    def size(self):
        return len(self.stack)


def measure_performance_stack(stack_class, operations):
    start_time = time.time()
    memory_before = sys.getsizeof(stack_class.stack)

    for i in range(operations):
        stack_class.push(i)

    for i in range(operations):
        stack_class.pop()

    end_time = time.time()
    memory_after = sys.getsizeof(stack_class.stack)

    time_taken = end_time - start_time
    memory_used = memory_after - memory_before

    return time_taken, memory_used


operations = 10000
list_stack = ListStack()
deque_stack = DequeStack()

list_stack_performance = measure_performance_stack(list_stack, operations)
deque_stack_performance = measure_performance_stack(deque_stack, operations)

list_stack_performance, deque_stack_performance
```

```python
def measure_individual_operations_and_memory_stack(stack_class, operations):
    push_times = []
    pop_times = []
    push_memory = []
    pop_memory = []

    for i in range(operations):
        start_time = time.time()
        stack_class.push(i)
        end_time = time.time()
        push_times.append(end_time - start_time)
        push_memory.append(sys.getsizeof(stack_class.stack))

    for i in range(operations):
        start_time = time.time()
        stack_class.pop()
        end_time = time.time()
        pop_times.append(end_time - start_time)
        pop_memory.append(sys.getsizeof(stack_class.stack))

    return push_times, pop_times, push_memory, pop_memory

operations = 200000


list_push_times, list_pop_times, list_push_memory, list_pop_memory = measure_individual_operations_and_memory_stack(list_stack, operations)
deque_push_times, deque_pop_times, deque_push_memory, deque_pop_memory = measure_individual_operations_and_memory_stack(deque_stack, operations)

plt.figure(figsize=(12, 12))

plt.subplot(2, 2, 1)
plt.plot(range(operations), list_push_times, color='blue', label='Время добавления в ListStack')
plt.plot(range(operations), deque_push_times, color='red', label='Время добавления в DequeStack')
plt.title('Время операций добавления в стек')
plt.xlabel('Номер операции')
plt.ylabel('Время (секунды)')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(range(operations), list_pop_times, color='blue', label='Время извлечения из ListStack')
plt.plot(range(operations), deque_pop_times, color='red', label='Время извлечения из DequeStack')
plt.title('Время операций извлечения из стека')
plt.xlabel('Номер операции')
plt.ylabel('Время (секунды)')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(range(operations), list_push_memory, color='blue', label='Использование памяти при добавлении в ListStack')
plt.plot(range(operations), deque_push_memory, color='red', label='Использование памяти при добавлении в DequeStack')
plt.title('Использование памяти во время добавления в стек')
plt.xlabel('Номер операции')
plt.ylabel('Использование памяти (байты)')
plt.legend()

# Использование памяти во время операций извлечения из стека
plt.subplot(2, 2, 4)
plt.plot(range(operations), list_pop_memory, color='blue', label='Использование памяти при извлечении из ListStack')
plt.plot(range(operations), deque_pop_memory, color='red', label='Использование памяти при извлечении из DequeStack')
plt.title('Использование памяти во время извлечения из стека')
plt.xlabel('Номер операции')
plt.ylabel('Использование памяти (байты)')
plt.legend()


plt.tight_layout()
plt.show()
```

Реализуйте класс Односвязный список. Найдите минимальный элемент в нем. Сравните с временем поиска минимального элемента при использовании списка.

```python
import time
import matplotlib.pyplot as plt

class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        if not self.head:
            self.head = Node(value)
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = Node(value)

    def find_min(self):
        if not self.head:
            return None

        min_val = self.head.value
        current = self.head
        while current:
            if current.value < min_val:
                min_val = current.value
            current = current.next
        return min_val

def measure_time_to_find_min_sll(sll, size):
    sll.head = None
    for i in range(size):
        sll.append(i)

    start_time = time.time()
    min_value = sll.find_min()
    end_time = time.time()
    return min_value, end_time - start_time

def measure_time_to_find_min_list(lst, size):
    start_time = time.time()
    min_value = min(lst[:size])
    end_time = time.time()
    return min_value, end_time - start_time

sizes = [100, 500, 1000, 5000, 10000, 15000, 250000]
times_sll = []
times_list = []

sll = SinglyLinkedList()
lst = list(range(250000))

for size in sizes:
    _, time_sll = measure_time_to_find_min_sll(sll, size)
    times_sll.append(time_sll)

    _, time_lst = measure_time_to_find_min_list(lst, size)
    times_list.append(time_lst)

plt.figure(figsize=(10, 6))
plt.plot(sizes, times_sll, label='Singly Linked List', marker='o')
plt.plot(sizes, times_list, label='Regular List', marker='x')
plt.title('Сравнение времени нахождения минимального элемента')
plt.xlabel('Размер списка')
plt.ylabel('Время (секунды)')
plt.legend()
plt.grid(True)
plt.show()

```

Реализуйте класс Двусвязный список.

```python
class DoubleNode:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, value):
        new_node = DoubleNode(value)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node

    def prepend(self, value):
        new_node = DoubleNode(value)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def delete(self, value):
        current = self.head
        while current:
            if current.value == value:
                if current.prev:
                    current.prev.next = current.next
                if current.next:
                    current.next.prev = current.prev
                if current == self.head:
                    self.head = current.next
                if current == self.tail:
                    self.tail = current.prev
                return
            current = current.next

    def __str__(self):
        values = []
        current = self.head
        while current:
            values.append(str(current.value))
            current = current.next
        return ' <-> '.join(values)

dll = DoublyLinkedList()
dll.append(1)
dll.append(2)
dll.append(3)
dll.prepend(0)
print("Doubly Linked List after appending and prepending:", dll)
dll.delete(2)
print("Doubly Linked List after deleting a value:", dll)
dll.prepend(10)
print("Doubly Linked List after deleting a value:", dll)
```
