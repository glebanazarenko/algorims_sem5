Есть список - это смесь числе и строк
написать программу, которая показыввет, сколько памяти выделяется на это

формируем два массива
и определяем, какой объем данных выделяется на эти два массива

```python
import sys

mixed_list = [1, 'a', 3.14, 'b', 'c', 7]

numbers = array.array('d') 
chars = array.array('u')

for item in mixed_list:
    if isinstance(item, (int, float)):
        numbers.append(item)
    elif isinstance(item, str) and len(item) == 1:
        chars.fromunicode(item)

print(numbers)
print(chars)
print('Memory size of numbers list:', sys.getsizeof(mixed_list), 'bytes')
print('Memory size of numbers list:', sys.getsizeof(numbers), 'bytes')
print('Memory size of strings list:', sys.getsizeof(chars), 'bytes')
```

Посторить график 
на вход подавем список разной длинны и в списке разное кол-в чисел 
как будет меняться память при разных кол-в элементах в списке

```python
import sys
import matplotlib.pyplot as plt
import numpy as np
import random

def generate_mixed_list(size):
    return [random.randint(0, 100) if random.choice([True, False]) else 'a' * random.randint(1, 10) for _ in range(size)]

sizes = np.arange(0, 10001, 1000)
memory_sizes = [sys.getsizeof(generate_mixed_list(size)) for size in sizes]

plt.figure(figsize=(10, 5))
plt.plot(sizes, memory_sizes, marker='o')
plt.title('Изменение объема памяти в зависимости от количества элементов в списке')
plt.xlabel('Количество элементов в списке')
plt.ylabel('Память (байты)')
plt.grid(True)
plt.show()
```

```python
import sys
import matplotlib.pyplot as plt
import numpy as np
import array

def calculate_memory_size(obj):
    return sys.getsizeof(obj)

def generate_number_array(size):
    return array.array('d', [i**10 for i in range(size)])

def generate_char_array(size):
    return array.array('u', ['a'] * size)

def generate_char_list(size):
    return ['a'] * size

sizes = np.arange(0, 10001, 1000) 
list_memory_sizes = [calculate_memory_size([i**10 for i in range(size)]) for size in sizes]
char_list_memory_sizes = [calculate_memory_size(generate_char_list(size)) for size in sizes]
number_array_memory_sizes = [calculate_memory_size(generate_number_array(size)) for size in sizes]
char_array_memory_sizes = [calculate_memory_size(generate_char_array(size)) for size in sizes]

plt.figure(figsize=(10, 5))
plt.plot(sizes, list_memory_sizes, marker='o', label='Список чисел')
plt.plot(sizes, char_list_memory_sizes, marker='^', label='Список символов')
plt.plot(sizes, number_array_memory_sizes, marker='s', label='Массив чисел')
plt.plot(sizes, char_array_memory_sizes, marker='d', label='Массив символов')
plt.title('Изменение объема памяти в зависимости от количества элементов')
plt.xlabel('Количество элементов')
plt.ylabel('Память (байты)')
plt.grid(True)
plt.legend()
plt.show()
```

Класс array.array(TypeCode [, инициализатор]) - новый массив, элементы которого ограничены TypeCode, и инициализатор, который должен быть списком, объектом, поддерживающим интерфейс буфера, или итерируемым объектом.

array.typecode - TypeCode символ, использованный при создании массива.

array.itemsize - размер в байтах одного элемента в массиве.

array.append(х) - добавление элемента в конец массива.

array.buffer_info() - кортеж (ячейка памяти, длина). Полезно для низкоуровневых операций.

array.byteswap() - изменить порядок следования байтов в каждом элементе массива. Полезно при чтении данных из файла, написанного на машине с другим порядком байтов.

array.count(х) - возвращает количество вхождений х в массив.

array.extend(iter) - добавление элементов из объекта в массив.

array.frombytes(b) - делает массив array из массива байт. Количество байт должно быть кратно размеру одного элемента в массиве.

array.fromfile(F, N) - читает N элементов из файла и добавляет их в конец массива. Файл должен быть открыт на бинарное чтение. Если доступно меньше N элементов, генерируется исключение EOFError , но элементы, которые были доступны, добавляются в массив.

array.fromlist(список) - добавление элементов из списка.

array.index(х) - номер первого вхождения x в массив.

array.insert(n, х) - включить новый пункт со значением х в массиве перед номером n. Отрицательные значения рассматриваются относительно конца массива.

array.pop(i) - удаляет i-ый элемент из массива и возвращает его. По умолчанию удаляется последний элемент.

array.remove(х) - удалить первое вхождение х из массива.

array.reverse() - обратный порядок элементов в массиве.

array.tobytes() - преобразование к байтам.

array.tofile(f) - запись массива в открытый файл.

array.tolist() - преобразование массива в список.

## Динамические массивы  <a class="anchor" id="динамические"></a>

```python
data = []
n = 33
for k in range(n):
    a = len(data) # количество элементов в списке
    b = sys.getsizeof(data) # текущий размер списка в байтах
    print(f'Length: {a:3d}; Size in bytes: {b:4d}')
    data.append(None) # добавляем элемент в список
```

```python
import sys
import matplotlib.pyplot as plt

data = []
n = 100
lengths = []  # Список для хранения количества элементов в списке на каждом шаге
sizes = []  # Список для хранения размера списка в байтах на каждом шаге

for k in range(n):
    a = len(data)  # количество элементов в списке
    b = sys.getsizeof(data)  # текущий размер списка в байтах
    print(f'Length: {a:3d}; Size in bytes: {b:4d}')
    lengths.append(a)
    sizes.append(b)
    data.append(None)  # добавляем элемент в список

# Построение графика
plt.figure(figsize=(10, 5))
plt.plot(lengths, sizes, marker='o')
plt.title('Изменение размера списка в памяти')
plt.xlabel('Количество элементов в списке')
plt.ylabel('Размер в байтах')
plt.grid(True)
plt.show()
```

пусть есть списко
он состоит из чисел (очень большой)
мы удаляем по одному элементу из списка (первый)
нужно оценить
1) как будет меняться память
2) как будет меняться время удаления

```python
import sys
import time
import matplotlib.pyplot as plt
import numpy as np

# Генерация большого списка чисел
n = 10000
data = list(range(n))

memory_sizes = []  # Для хранения размеров списка в байтах
time_deltas = []   # Для хранения времени удаления элементов в секундах

for _ in range(n):
    memory_sizes.append(sys.getsizeof(data))
    
    start_time = time.perf_counter()
    del data[0]
    end_time = time.perf_counter()
    
    time_deltas.append(end_time - start_time)

# Создание массива индексов для оси X
indices = np.arange(n)

# Построение графиков
fig, ax1 = plt.subplots(figsize=(10, 5))

color = 'tab:red'
ax1.set_xlabel('Количество удаленных элементов')
ax1.set_ylabel('Память (байты)', color=color)
ax1.plot(indices, memory_sizes, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Время (секунды)', color=color)
ax2.plot(indices, time_deltas, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title('Изменение памяти и времени удаления элементов из списка')
plt.show()
```

__Асимптотическая эффективность операций для списка (кортежа) (list/tuple), не меняющих содержимое структуры данных.__

Идентификаторы объектов: data или data1 и data2, и соответственно их длины: $n$, $n_1$ и $n_2$. $k$ обозначает индекс первого вхождения (слева) искомого элемента, в случае отсутствия элемента $k=n$. При сравнении двух последовательностей k показывает самый левый элемент, на котором последовательности не согласуются или $k = min(n_1, n_2)$.
